```python
import pandas as pd 
import numpy as np 
```

```python
data=pd.read_csv("datasets/cardio_train.csv", sep=";")
data
```

```python
data.shape
```

```python
data.describe()
```

```python
data.info()
```

```python
#converting height from cm into m

data["height"]=data["height"]/float(100)
data.head(5)
```

```python
#calculating the bmi feature

data["bmi"]=data["weight"]/(data["height"]**2)
data.head(5)
```

```python
#converting age from days to years
data["age"]=data["age"]//365
```

```python
data.head(5)
```

```python
data['age'].describe()
```

```python
X=data[['gender', 'age', 'smoke', 'alco', 'active', 'bmi']]
Y=data['cardio']
X.head(10)
```

```python
X.info()
```

```python
X["male"]=[1 if x==2 else 0 for x in X["gender"]]
X["female"]=[1 if x==1 else 0 for x in X["gender"]]
X.head(10)
```

```python
X.drop("gender",axis=1, inplace=True)
X.head(10)
```

```python
#trying with lLogisticRegression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test =train_test_split(X,Y, random_state=42, test_size=0.2)
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
#trying with decision tree
```
