```python
X=data[["Age", "BMI"]]
Y=data["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
lg=LogisticRegression(class_weight="balanced", random_state=42)
dt=DecisionTreeClassifier(max_depth=2, class_weight="balanced", random_state=42, criterion="gini")
rfc=RandomForestClassifier(class_weight="balanced", random_state=42)
```

```python
lg.fit(x_train, y_train)
dt.fit(x_train, y_train) 
rfc.fit(x_train, y_train)  
```

```python
lg_pred1=lg.predict(x_test)
dt_pred1=dt.predict(x_test) 
rfc_pred1=rfc.predict(x_test) 
```

```python
def get_result(y_test,y_pred):
    print("Classification Report:\n\n")
    print(classification_report(y_test, y_pred))
    print("\n\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("\n\naccuracy scrore= ", accuracy_score(y_test, y_pred))
    
```

```python
print("Logistic Regression: \n")
get_result(y_test,lg_pred1)
print("\n\nDecision Tree: \n")
get_result(y_test,dt_pred1)
print("\n\nRandom Forest: \n")
get_result(y_test,rfc_pred1)
```
